```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('Expanded_data_with_more_features.csv')
df.drop(['Unnamed: 0'], axis=1)
df['EthnicGroup'].unique()
# df = df['EthnicGroup'].fillna('group C')
df[['EthnicGroup']] = df[['EthnicGroup']].fillna('group C')

df['ParentEduc'].unique()
df[['ParentEduc']] = df[['ParentEduc']].fillna('some college')

df['TestPrep'].unique()
df[['TestPrep']] = df[['TestPrep']].fillna('nan')

df['ParentMaritalStatus'].unique()
df[['ParentMaritalStatus']] = df[['ParentMaritalStatus']].fillna('married')

df['PracticeSport'].unique()
df[['PracticeSport']] = df[['PracticeSport']].fillna('sometimes')

df['IsFirstChild'].unique()
df[['IsFirstChild']] = df[['IsFirstChild']].fillna('yes')

df['NrSiblings'].unique()
df[['NrSiblings']] = df[['NrSiblings']].fillna(1.0)

df['TransportMeans'].unique()
df[['TransportMeans']] = df[['TransportMeans']].fillna('other')

df['WklyStudyHours'].unique()
df[['WklyStudyHours']] = df[['WklyStudyHours']].fillna('5  - 10')
df.isnull().sum()
```

```python
label_encoder = LabelEncoder()
encoded_data = label_encoder.fit_transform(df['Gender'])
df['Gender'] = encoded_data

encoded_data = label_encoder.fit_transform(df['EthnicGroup'])
df['EthnicGroup'] = encoded_data

encoded_data = label_encoder.fit_transform(df['ParentEduc'])
df['ParentEduc'] = encoded_data

encoded_data = label_encoder.fit_transform(df['LunchType'])
df['LunchType'] = encoded_data

encoded_data = label_encoder.fit_transform(df['TestPrep'])
df['TestPrep'] = encoded_data

encoded_data = label_encoder.fit_transform(df['ParentMaritalStatus'])
df['ParentMaritalStatus'] = encoded_data

encoded_data = label_encoder.fit_transform(df['PracticeSport'])
df['PracticeSport'] = encoded_data

encoded_data = label_encoder.fit_transform(df['IsFirstChild'])
df['IsFirstChild'] = encoded_data

encoded_data = label_encoder.fit_transform(df['NrSiblings'])
df['NrSiblings'] = encoded_data

encoded_data = label_encoder.fit_transform(df['TransportMeans'])
df['TransportMeans'] = encoded_data

encoded_data = label_encoder.fit_transform(df['WklyStudyHours'])
df['WklyStudyHours'] = encoded_data

encoded_data = label_encoder.fit_transform(df['MathScore'])
df['MathScore'] = encoded_data

encoded_data = label_encoder.fit_transform(df['ReadingScore'])
df['ReadingScore'] = encoded_data

encoded_data = label_encoder.fit_transform(df['WritingScore'])
df['WritingScore'] = encoded_data

df
```
