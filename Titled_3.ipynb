```python
import pandas as pd


df = pd.read_csv('results.csv')
lst = ['Hindi', 'English', 'Science', 'Maths', 'History',
       'Geograpgy', 'Results']
df = df[lst]
df
```

```python
X = df[lst[:-1]]
X #баллы за задание
```

```python
y = df['Results']
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression().fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
y_pred
```

```python
#accurace = TP + TN / TP + TN + FP + FN
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
#presision = TP / TP + FP
from sklearn.metrics import precision_score
precision_score(y_test, y_pred)
```

```python
#recall = TP / TP + FN
from sklearn.metrics import recall_score
recall_score(y_test, y_pred)
```

```python
#F = 2 * presision + recall / presision + recall
from sklearn.metrics import f1_score
f1_score(y_test, y_pred)
```

```python
def mark_predict(hindi, english, science, maths, history, geography):
  marks = [[hindi, english, science, maths, history, geography]]
  df2 = pd.DataFrame(marks, columns=lst[:-1])
  result = clf.predict(df2)[0]
  return result
```
