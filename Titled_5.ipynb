```python
import math

def distance(p1, p2):
  x1 = p1[0]
  x2 = p2[0]
  y1 = p1[1]
  y2 = p2[1]
  return math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
```

```python
def update_clusters(data, clusters, centers):
  for i in range(len(data)):
    dist = []
    for coor in centers:
      d = distance(data[i], coor)
      dist.append(d)
    mn = min(dist)
    k = dist.index(mn)
    clusters[i] = k
  return clusters
```
