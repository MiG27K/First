```python
def f(x, y):
  return 2 * x + y - 10
```

```python
def sigma(z):
  return 1 / (1 + np.exp(-z))
```

```python
sigma(f(1,3)), sigma(f(4, 3)), sigma(f(1, 7)), sigma(f(5, 2)), sigma(f(3, 6)) #если больше 0.5 - в 1 класс, меньше - во 2
```
